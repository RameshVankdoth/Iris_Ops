```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_absolute_error, accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
iris = load_iris()
lr = LogisticRegression()
```

```python
x = iris.data
y = iris.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30)
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
mse = mean_absolute_error(y_test, y_pred)
mse
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
plt.scatter(y_test, y_pred)
```

```python
plt.plot(y_test, y_pred, color='red')
plt.xlabel('petal length')
plt.ylabel('target')
plt.title('Linear Regression with Rounded Predictions')
plt.show()
```

```python
cmd = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred))
cmd.plot()
```
